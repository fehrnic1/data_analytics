# Importing and preparing supermarkets data

## Libraries and settings

```python
# Libraries
import os
import fnmatch
import pandas as pd

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Get current working directory
print('Current working directory:', os.getcwd())

# Show .json files in the current working directory
flist = fnmatch.filter(os.listdir('.'), '*.json')
for i in flist:
    print(i)
```

## Importing data

```python
# Read the data to a pandas data frame
df1 = pd.read_json('supermarkets.json', encoding='utf-8')
df1.head(5)
```

## Count number of rows and columns in the data frame

```python
# Dimension (rows, columns)
print('Dimension:', df1.shape)

# Number of rows
print('Number of rows:', df1.shape[0])

# Number of columns
print('Number of columns:', df1.shape[1])
```

## Column 'tags' is a pandas Series with dictionaries -> change to data frame

```python
# Type of the first item of column tags
print(type(df1.tags))
print(type(df1.tags[0]))

# Content of the first item of column tags
print(df1.tags[0].keys())

# Change to data frame
df2 = pd.DataFrame.from_records(df1.tags)
df2 = df2[['brand', 'shop', 'addr:city', 'addr:street', 'addr:housenumber', 'addr:postcode']]

# Rename selected columns
df2 = df2.rename(columns={'addr:city': 'city',
                          'addr:street':'street',
                          'addr:housenumber': 'housenumber',
                          'addr:postcode': 'postcode'})

# Show first records of data frame
df2.head()
```

## Merge df1 and df2

```python
# Merge df and df2
df = pd.merge(df1[['type', 'id', 'lat', 'lon']], 
              df2[['brand', 'shop', 'city', 'street', 'housenumber', 'postcode']],
              left_index=True, 
              right_index=True)
df.head(5)
```

## Count and identify the number of missing values (if any)

```python
# Count missing values
print(pd.isna(df).sum())

# Identify rows with missing values, e.g.:
df.loc[pd.isna(df['city'])]
```

```python
# Filter data frame for brand 'Migros' and city 'Zürich'
df_filtered1 = df.loc[(df['brand'] == 'Migros') & (df['city'] == 'Zürich')]
print(df_filtered1[['id', 'brand', 'city', 'street', 'housenumber', 'postcode']])

# Filter data frame for brand 'Coop' in cities 'Zürich', 'Bern', 'Basel'
df_filtered2 = df.loc[(df['brand'] == 'Coop') & (df['city'].isin(['Zürich', 'Bern', 'Basel']))]
print(df_filtered2[['id', 'brand', 'city', 'street', 'housenumber', 'postcode']])

# Filter data frame for entries with available brand, city, house number and postcode
df_filtered3 = df.loc[(df['brand'].notnull()) & (df['city'].notnull()) & (df['housenumber'].notnull()) & (df['postcode'].notnull())]
print(df_filtered3[['id', 'brand', 'city', 'street', 'housenumber', 'postcode']])

# Include opening hours as additional variable in the data frame. Then filter for supermarkets with opening hours information available.
df_final = pd.merge(df, df1[['tags']], left_index=True, right_index=True)
df_final['opening_hours'] = df_final['tags'].apply(lambda x: x.get('opening_hours') if isinstance(x, dict) else None)
df_final_filtered = df_final.loc[df_final['opening_hours'].notnull()]
print(df_final_filtered[['id', 'brand', 'city', 'street', 'housenumber', 'postcode', 'opening_hours']]) 
```

## Count and identify duplicated values (if any)

```python
# Count duplicated values
print(df.duplicated().sum())

# Identify rows with duplicated values, e.g.:
df[df[['id']].duplicated()]
```

## Get data types of all variables

```python
# Get data types (note that in pandas, a string is referred to as 'object')
df.dtypes
```

### Save data to file

```python
df.to_csv('supermarkets_data_prepared.csv', 
          sep=",", 
          encoding='utf-8',
          index=False)
```
